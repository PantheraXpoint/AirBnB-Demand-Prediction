```python
def merge_dfs(road, hf, air):
    hf.rename(columns={'기준일ID':'Reporting Month', '행정동코드':'ADM_NM'}, inplace=True)
    air_new = air.copy()
    air_new['Reporting Month'] = pd.to_datetime(air['Reporting Month'], format='%Y-%m-%d').dt.strftime('%Y%m')
    air_new['Reporting Month'] = air_new['Reporting Month'].astype(str)  # 우선 문자열로 변환
    hf['Reporting Month'] = hf['Reporting Month'].astype(str)  # 우선 문자열로 변환

    # 도로 크기 맞추기
    months = [f"{year}{str(month).zfill(2)}" for year in range(2017, 2023) for month in range(1, 13)][:67]  # Create 67 months (YYMM format)
    road_expanded = road.loc[road.index.repeat(len(months))].reset_index(drop=True)  # Repeat each row 67 times
    road_expanded['Reporting Month'] = months * len(road)  # Assign 67 months for each 'ADM_NM'
    
    df = pd.merge(hf, air_new, on=['ADM_NM', 'Reporting Month'], how='inner')  # 이게 문제..
    #print(df.shape)
    
    df = pd.merge(df, road_expanded, on=['ADM_NM', 'Reporting Month'], how='inner')
    #print(df.shape)
    return df
```

```python
def merge_dfs2(hf, air):
    hf.rename(columns={'기준일ID':'Reporting Month', '행정동코드':'ADM_NM'}, inplace=True)
    air_new = air.copy()
    air_new['Reporting Month'] = pd.to_datetime(air['Reporting Month'], format='%Y-%m-%d').dt.strftime('%Y%m')
    air_new['Reporting Month'] = air_new['Reporting Month'].astype(str)  # 우선 문자열로 변환
    hf['Reporting Month'] = hf['Reporting Month'].astype(str)  # 우선 문자열로 변환

    
    df = pd.merge(hf, air_new, on=['ADM_NM', 'Reporting Month'], how='inner')
    #print(df.shape)
    
    return df
```

```python
hf_air_gcn = merge_dfs2(hf_gcn, air_gcn)
hf_air_gat = merge_dfs2(hf_gat, air_gat)

print(hf_air_gcn.shape)
print(hf_air_gat.shape)
```
